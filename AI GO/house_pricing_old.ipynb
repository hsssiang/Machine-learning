### 將all_data的橫坐標與縱座標（緯度跟經度）從TWD97 二度分帶座標值轉換為 WGS84經緯度坐標值

```python
# 將all_data的橫坐標與縱座標（緯度跟經度）從TWD97 二度分帶座標值轉換為 WGS84經緯度坐標值
import pyproj

def TWD97_to_lnglat(x, y):
    projTWD97 = pyproj.Proj("+proj = tmerc +lat_0 = 0 +lon_0 = 121 +k = 0.9999 +x_0 = 250000 +y_0 = 0 +ellps = GRS80 +units = m +no_defs")
    projWGS84 = pyproj.Proj("+proj = longlat +ellps = WGS84 +datum = WGS84 +no_defs")
    lng, lat = pyproj.transform(projTWD97, projWGS84, x, y)
    return lng, lat

all_data['lng'], all_data['lat'] = TWD97_to_lnglat(all_data['橫坐標'], all_data['縱坐標'])

#印出轉換後結果
all_data[['縱坐標', '橫坐標', 'lat', 'lng']]
```

```python
# 檢查all_data的'lat' 跟 'lng'是否有異常值 lat > 27 | lat < 21 | lng > 125 | lng < 116 | dropna
all_data = all_data.drop(all_data[(all_data['lat'] > 27) | (all_data['lat'] < 21) | (all_data['lng'] > 125) | (all_data['lng'] < 116)].index).dropna()
all_data.shape
```

### 將捷運站距離及數量作為特徵
### 使用了台中市政府資料開放平臺所提供的台中捷運綠線資訊
### MRT.shape: (257, 7)
### nothing is dropped

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np

# 根據捷運站點資料和all_data的lat lng，找出all_data各比資料最近的n個捷運站距離，並以['捷運站距離和倒數']作為all_data新的欄位
path1 = './30_Training Dataset_V2/external_data/捷運站點資料.csv'
path2 = './30_Training Dataset_V2/external_data/臺中捷運綠線車站資訊.csv'
MRT1 = pd.read_csv(path1)
MRT2 = pd.read_csv(path2)
# 將MRT2的'緯度'改為'lat'、'經度'改為'lng'
MRT2 = MRT2.rename(columns = {'緯度':'lat', '經度':'lng'})

MRT = pd.concat([MRT1, MRT2])
MRT

print('before drop')
print(MRT.shape)
print(MRT.info())

# drop lat > 27 | lat < 21 | lng > 125 | lng < 116 | dropna
MRT = MRT.drop(MRT[(MRT['lat'] > 27) | (MRT['lat'] < 21) | (MRT['lng'] > 125) | (MRT['lng'] < 116)].index).dropna(subset = ['lat', 'lng'])

print('\nafter drop')
print(MRT.shape)
print(MRT.info())

# 設定要找出最近鄰居的數量n
n_neighbors = 1

# 計算最近n鄰居
knn = NearestNeighbors(n_neighbors = n_neighbors, algorithm = 'ball_tree').fit(MRT[['lat', 'lng']])
distances, indices = knn.kneighbors(all_data[['lat', 'lng']])

# 距離開根號和的倒數
#distances = np.sqrt(distances)
distances_sum = np.sum(distances, axis = 1)
distances_sum = 1 / distances_sum
#distances_sum_log = np.log2(distances_sum)

# 將小於1的設為0
#distances_sum[distances_sum < 1] = 0
all_data['捷運站距離和倒數'] = distances_sum
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
import math
def twd97_to_lonlat( x = 174458.0, y = 2525824.0 ):
    a = 6378137
    b = 6356752.314245
    long_0 = 121 * math.pi / 180.0
    k0 = 0.9999
    dx = 250000
    dy = 0
    
    e = math.pow((1-math.pow(b, 2)/math.pow(a,2)), 0.5)
    
    x -= dx
    y -= dy
    
    M = y / k0
    
    mu = M / ( a*(1-math.pow(e, 2)/4 - 3*math.pow(e,4)/64 - 5 * math.pow(e, 6)/256))
    e1 = (1.0 - pow((1   - pow(e, 2)), 0.5)) / (1.0 +math.pow((1.0 -math.pow(e,2)), 0.5))
    
    j1 = 3*e1/2-27*math.pow(e1,3)/32
    j2 = 21 * math.pow(e1,2)/16 - 55 * math.pow(e1, 4)/32
    j3 = 151 * math.pow(e1, 3)/96
    j4 = 1097 * math.pow(e1, 4)/512
    
    fp = mu + j1 * math.sin(2*mu) + j2 * math.sin(4* mu) + j3 * math.sin(6*mu) + j4 * math.sin(8* mu)
    
    e2 = math.pow((e*a/b),2)
    c1 = math.pow(e2*math.cos(fp),2)
    t1 = math.pow(math.tan(fp),2)
    r1 = a * (1-math.pow(e,2)) / math.pow( (1-math.pow(e,2)* math.pow(math.sin(fp),2)), (3/2))
    n1 = a / math.pow((1-math.pow(e,2)*math.pow(math.sin(fp),2)),0.5)
    d = x / (n1*k0)
    
    q1 = n1* math.tan(fp) / r1
    q2 = math.pow(d,2)/2
    q3 = ( 5 + 3 * t1 + 10 * c1 - 4 * math.pow(c1,2) - 9 * e2 ) * math.pow(d,4)/24
    q4 = (61 + 90 * t1 + 298 * c1 + 45 * math.pow(t1,2) - 3 * math.pow(c1,2) - 252 * e2) * math.pow(d,6)/720
    lat = fp - q1 * (q2 - q3 + q4)
    
    
    q5 = d
    q6 = (1+2*t1+c1) * math.pow(d,3) / 6
    q7 = (5 - 2 * c1 + 28 * t1 - 3 * math.pow(c1,2) + 8 * e2 + 24 * math.pow(t1,2)) * math.pow(d,5) / 120
    lon = long_0 + (q5 - q6 + q7) / math.cos(fp)
    
    lat = (lat*180) / math.pi
    lon = (lon*180) / math.pi
    return lat, lon

industry_lat = list()
industry_lng = list()

# 根據捷運站點資料和all_data的lat lng，找出all_data各比資料最近的n個捷運站距離，並以['捷運站距離和倒數']作為all_data新的欄位

path1 = '30_Training Dataset_V2/external_data/科技工業園區.csv'
Industry1 = pd.read_csv(path1)

# 將MRT2的'緯度'改為'lat'、'經度'改為'lng'
Industry1 = Industry1.rename(columns = {'X':'lat', 'Y':'lng'})

print('before drop')
print(Industry1.shape)
print(Industry1.info())
Industry1 = Industry1.drop( ['縣市','名稱','開發依','開發情','管理單','面積','類別'], axis=1 )
# drop lat > 27 | lat < 21 | lng > 125 | lng < 116 | dropna
industry_lat_twd97 = Industry1['lat'].values.tolist()
industry_lng_twd97 = Industry1['lng'].values.tolist()
for i,j in zip(industry_lat_twd97, industry_lng_twd97):
    a,b = twd97_to_lonlat( i, j )
    industry_lat.append(a)
    industry_lng.append(b)
Industry_final = { 'lat' : industry_lat , 'lng' : industry_lng }
Industry_final = pd.DataFrame(Industry_final)
Industry_final = Industry_final.drop(Industry_final[(Industry_final['lat'] > 27) | (Industry_final['lat'] < 21) | (Industry_final['lng'] > 125) | (Industry_final['lng'] < 116)].index).dropna(subset = ['lat', 'lng'])

print('\nafter drop')
print(Industry_final.shape)
print(Industry_final.info())

# 設定要找出最近鄰居的數量n
n_neighbors = 3

# 計算最近n鄰居
knn = NearestNeighbors(n_neighbors = n_neighbors, algorithm = 'ball_tree').fit(Industry_final[['lat', 'lng']])
distances, indices = knn.kneighbors(all_data[['lat', 'lng']])

# 距離開根號和的倒數
#distances = np.sqrt(distances)
distances_sum = np.sum(distances, axis = 1)
distances_sum = 1 / distances_sum
#distances_sum_log = np.log2(distances_sum)

# 將小於1的設為0
#distances_sum[distances_sum < 1] = 0
all_data['工業園區距離和倒數'] = distances_sum
```
