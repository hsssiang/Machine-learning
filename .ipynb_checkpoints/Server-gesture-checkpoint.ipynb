```python
import cv2
import mediapipe as mp
import math
import time
from tkinter import *
import tkinter as tk
from PIL  import Image, ImageTk
import random
import statistics
import  socket

ip = '192.168.56.1'
port = 5000
c = socket.socket(socket.AF_INET,socket.SOCK_STREAM)
c.connect((ip,port))

cap = cv2.VideoCapture(0)
mpHands = mp.solutions.hands
hands = mpHands.Hands(min_detection_confidence=0.5, min_tracking_confidence=0.5)
mpDraw = mp.solutions.drawing_utils
handLmsStyle = mpDraw.DrawingSpec(color=(0, 0, 255), thickness=3)
handConStyle = mpDraw.DrawingSpec(color=(0, 255, 0), thickness=5)
pTime = 0
cTime = 0
text = 0

# 根據兩點的座標，計算角度
def vector_2d_angle(v1, v2):
    v1_x = v1[0]
    v1_y = v1[1]
    v2_x = v2[0]
    v2_y = v2[1]
    try:
        angle_= math.degrees(math.acos((v1_x*v2_x+v1_y*v2_y)/(((v1_x**2+v1_y**2)**0.5)*((v2_x**2+v2_y**2)**0.5))))
    except:
        angle_ = 180
    return angle_

# 根據傳入的 21 個節點座標，得到該手指的角度
def hand_angle(hand_):
    angle_list = []
    # thumb 大拇指角度
    angle_ = vector_2d_angle(
        ((int(hand_[0][0])- int(hand_[2][0])),(int(hand_[0][1])-int(hand_[2][1]))),
        ((int(hand_[3][0])- int(hand_[4][0])),(int(hand_[3][1])- int(hand_[4][1])))
        )
    angle_list.append(angle_)
    # index 食指角度
    angle_ = vector_2d_angle(
        ((int(hand_[0][0])-int(hand_[6][0])),(int(hand_[0][1])- int(hand_[6][1]))),
        ((int(hand_[7][0])- int(hand_[8][0])),(int(hand_[7][1])- int(hand_[8][1])))
        )
    angle_list.append(angle_)
    # middle 中指角度
    angle_ = vector_2d_angle(
        ((int(hand_[0][0])- int(hand_[10][0])),(int(hand_[0][1])- int(hand_[10][1]))),
        ((int(hand_[11][0])- int(hand_[12][0])),(int(hand_[11][1])- int(hand_[12][1])))
        )
    angle_list.append(angle_)
    # ring 無名指角度
    angle_ = vector_2d_angle(
        ((int(hand_[0][0])- int(hand_[14][0])),(int(hand_[0][1])- int(hand_[14][1]))),
        ((int(hand_[15][0])- int(hand_[16][0])),(int(hand_[15][1])- int(hand_[16][1])))
        )
    angle_list.append(angle_)
    # pink 小拇指角度
    angle_ = vector_2d_angle(
        ((int(hand_[0][0])- int(hand_[18][0])),(int(hand_[0][1])- int(hand_[18][1]))),
        ((int(hand_[19][0])- int(hand_[20][0])),(int(hand_[19][1])- int(hand_[20][1])))
        )
    angle_list.append(angle_)
    
    f12 = math.hypot((int(hand_[4][0])- int(hand_[8][0])),(int(hand_[4][1])- int(hand_[8][1])))
    angle_list.append(f12)   
    
    f23 = math.hypot((int(hand_[8][0])- int(hand_[12][0])),(int(hand_[8][1])- int(hand_[12][1])))
    angle_list.append(f23)
    
    f34 = math.hypot((int(hand_[12][0])- int(hand_[16][0])),(int(hand_[12][1])- int(hand_[16][1])))
    angle_list.append(f34)
    
    f45 = math.hypot((int(hand_[16][0])- int(hand_[20][0])),(int(hand_[16][1])- int(hand_[20][1])))
    angle_list.append(f45)
    
    print(f12,f23,f34,f45)
    return angle_list

# 根據手指角度的串列內容，返回對應的手勢名稱
def hand_pos(finger_angle):
    f1 = finger_angle[0]   # 大拇指角度
    f2 = finger_angle[1]   # 食指角度
    f3 = finger_angle[2]   # 中指角度
    f4 = finger_angle[3]   # 無名指角度
    f5 = finger_angle[4]   # 小拇指角度
    
    f12 = finger_angle[5]
    f23 = finger_angle[6]
    f34 = finger_angle[7]
    f45 = finger_angle[8]

    # 小於 50 表示手指伸直，大於等於 50 表示手指捲縮
    if f1>=50 and f2<50 and f3>=50 and f4>=50 and f5>=50:
        return 'A' #食
    elif f1>=50 and f2<50 and f3<50 and f4>=50 and f5>=50 and f23<50:
        return 'B' #食中並
    elif f1>=50 and f2<50 and f3<50 and f4>=50 and f5>=50 and f23>=50:
        return 'C' #食中開
    elif f1>=50 and f2<50 and f3<50 and f4<50 and f5>=50 and f23<50 and f34<50:
        return 'D' #食中無病病
    elif f1>=50 and f2<50 and f3<50 and f4<50 and f5>=50 and f23<50 and f34>=50:
        return 'E' #食中無並開
    elif f1>=50 and f2<50 and f3<50 and f4<50 and f5>=50 and f23>=50 and f34<50:
        return 'F' #食中無開並
    elif f1>=50 and f2<50 and f3<50 and f4<50 and f5>=50 and f23>=50 and f34>=50:
        return 'G' #食中無開開
    elif f1>=50 and f2<50 and f3<50 and f4<50 and f5<50 and f23<50 and f34<50 and f45<50:
        return 'H' #食中無小病病病
    elif f1>=50 and f2<50 and f3<50 and f4<50 and f5<50 and f23<50 and f34<50 and f45>=50:
        return 'I' #食中無小病病開
    elif f1>=50 and f2<50 and f3<50 and f4<50 and f5<50 and f23<50 and f34>=50 and f45<50:
        return 'J' #食中無小病開病
    elif f1>=50 and f2<50 and f3<50 and f4<50 and f5<50 and f23>=50 and f34<50 and f45<50:
        return 'K' #食中無小開病病
    elif f1>=50 and f2<50 and f3<50 and f4<50 and f5<50 and f23>=50 and f34>=50 and f45>=50:
        return 'L' #食中無小開開開
    elif f1<50 and f2<50 and f3<50 and f4<50 and f5<50:
        return 'M'
    elif f1<50 and f2>=50 and f3>=50 and f4>=50 and f5<50:
        return 'N'
    elif f1<50 and f2<50 and f3>=50 and f4>=50 and f5>=50:
        return 'O'
    elif f1<50 and f2<50 and f3<50 and f4>=50 and f5>=50:
        return 'P'
    elif f1<50 and f2<50 and f3<50 and f4<50 and f5>=50:
        return 'Q'
    elif f1>=50 and f2>=50 and f3>=50 and f4>=50 and f5>=50:
        return 'R'
    elif f1<50 and f2>=50 and f3>=50 and f4>=50 and f5>=50:
        return 'S'
    elif f1>=50 and f2>=40 and f3<50 and f4<50 and f5<50:
        return 'T'
    elif f1<50 and f2>=40 and f3<50 and f4<50 and f5<50:
        return 'T'
    elif f1>=50 and f2>=50 and f3>=50 and f4>=50 and f5<50:
        return 'U'
    elif f1<50 and f2<50 and f3>=50 and f4>=50 and f5<50:
        return 'V'
    elif f1>=50 and f2>=50 and f3>=50 and f4<50 and f5<50:
        return 'W'
    elif f1<50 and f2>=50 and f3<50 and f4>=50 and f5>=50:
        return 'X'
    elif f1>=50 and f2<50 and f3>=50 and f4>=50 and f5<50:
        return 'Y'
    elif f1<50 and f2>=50 and f3<50 and f4>=50 and f5<50:
        return 'Z'
    elif f1<50 and f2>=50 and f3>=50 and f4<50 and f5<50:
        return 'ERROR'
    elif f1>=50 and f2<50 and f3<50 and f4>=50 and f5<50:
        return 'SWITCH'
    elif f1<50 and f2<50 and f3<50 and f4>=50 and f5<50:
        return 'CANCEL'
    elif f1<50 and f2<50 and f3>=50 and f4<50 and f5<50:
        return 'READY'
    else:
        return ''

fontFace = cv2.FONT_HERSHEY_SIMPLEX  # 印出文字的字型
lineType = cv2.LINE_AA               # 印出文字的邊框

# mediapipe 啟用偵測手掌
with mpHands.Hands( #啟用偵測手掌
    model_complexity=0, min_detection_confidence=0.5, min_tracking_confidence=0.5) as hands:
    if not cap.isOpened():
        print("無法開啟WebCam")
        exit()
    w, h = 480, 360
    while True:
        ret, img = cap.read()
        if ret:
            imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            result = hands.process(imgRGB)

            #print(result.multi_hand_landmarks)

            imgHeight = img.shape[0]
            imgWidth = img.shape[1]

            if result.multi_hand_landmarks:
                for handLms in result.multi_hand_landmarks:
                    mpDraw.draw_landmarks(img, handLms, mpHands.HAND_CONNECTIONS, handLmsStyle, handConStyle)
                    for i, lm in enumerate(handLms.landmark):
                        xPos = int(lm.x * imgWidth)
                        yPos = int(lm.y * imgHeight)

                        #cv2.putText(img, str(i), (xPos-25, yPos+5), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0, 0, 255), 2)

                        #if i == 4:
                        #    cv2.circle(img, (xPos, yPos), 20, (166, 56, 56), cv2.FILLED)
                        #print(i, xPos, yPos)
            img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # 轉換色彩
            results = hands.process(img2)  # 偵測手勢
            if results.multi_hand_landmarks:
                for hand_landmarks in results.multi_hand_landmarks:
                    points = []  # 記錄手指節點座標的串列
                    for i in hand_landmarks.landmark:  # 將21個節點換算成座標
                        x = i.x * w
                        y = i.y * h
                        points.append((x, y))
                    if points:
                        angles = hand_angle(points)  # 計算手指角度
                        text = hand_pos(angles)  # 輸入手指的角度取得手勢名稱
                        cv2.putText(img, text, (30, 120), fontFace, 3, (255, 0, 0), 5, lineType)  # 印出手勢所表示意義
            list1=[]
            list1.append(text)
            list2=[' ',' ']
            maxlabel = max(list1,key=list1.count)
            list2.append(maxlabel) 
            
            msg = list2[-1]
            c.send(str(msg).encode('utf-8'))
                        
            cTime = time.time()
            fps = 1/(cTime-pTime)
            pTime = cTime
            cv2.putText(img, f"FPS : {int(fps)}", (30, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 3)
            

        cv2.imshow('img', img)
        if cv2.waitKey(1) == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```
